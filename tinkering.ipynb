```python
# Initialize and train the Gradient Boosting models with class weights
print('Begin GBDT Implementation with Class Weights...')

# GBDT with max_depth = 2
gbdt_2_model = GradientBoostingClassifier(n_estimators=200, max_depth=6, random_state=1234)
sample_weight_2 = np.where(y_train == 0, 1, class_weight)
gbdt_2_model.fit(X_train, y_train, sample_weight=sample_weight_2)
gbdt_2_predictions = gbdt_2_model.predict_proba(X_test)[:, 1].tolist()

print('End')
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import make_scorer, roc_auc_score
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_curve
import numpy as np

# Define class weights
class_weight = len(y_train) / (2 * sum(y_train))

# Create a custom scoring function based on the mean ROC AUC with class weights
def custom_roc_auc_score(y_true, y_pred):
    fpr, tpr, _ = roc_curve(y_true, y_pred)
    roc_auc = roc_auc_score(y_true, y_pred)
    weighted_roc_auc = roc_auc * (1 - fpr) * class_weight
    return np.mean(weighted_roc_auc)  # Calculate the mean of the ROC AUC scores

# Create a GradientBoostingClassifier
gbdt_model = GradientBoostingClassifier(random_state=1234)

# Define the hyperparameters to tune
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [2, 4, 6, 8],
    # Add other hyperparameters to tune
}

# Perform cross-validation with Grid Search and custom scoring
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=1234)

grid_search = GridSearchCV(gbdt_model, param_grid, cv=cv, scoring=make_scorer(custom_roc_auc_score, needs_proba=True), verbose=2, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_

# Train the model with the best hyperparameters
best_gbdt_model = GradientBoostingClassifier(**best_params, random_state=1234)
best_gbdt_model.fit(X_train, y_train)

# Get ROC AUC scores for different thresholds on the best model
y_scores = best_gbdt_model.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_scores)

# Print the best hyperparameters and the best ROC AUC score
print("Best Hyperparameters:", best_params)
print("Best Weighted ROC AUC Score:", grid_search.best_score_)

# Optionally, you can plot the ROC curve
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()


```
